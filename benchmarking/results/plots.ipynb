```python
fig, ax = plt.subplots(figsize=(5, 5), dpi=600)

x = ['Vanilla', 'CoT', 'Self-Learning', 'Self-Learning + Memory'] 
y = [(43/102)*100, (49/102)*100, (65/102)*100, (67/93)*100]
y = [round(i, 2) for i in y]

sns.barplot(x, y, palette='Blues_d')

# ax.set_xticklabels(x, rotation=25, fontsize=8)
ax.set_xlabel("")
ax.set_ylabel("Accuracy \%", fontsize=9)
ax.set_ylim(0, 100)

ax.bar_label(ax.containers[0], fontsize=7)

plt.title('MMLU College Physics Test', fontsize=9)

for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(9) 

# plt.savefig('img.png')
plt.show()
```
